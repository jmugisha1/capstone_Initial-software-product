```python
import pickle
from sentence_transformers import SentenceTransformer
from tensorflow.keras.models import load_model
```

```python
# Load everything once
model = load_model('/content/drive/MyDrive/capstone/model/disease_classifier.h5')

# Load label encoder
with open('/content/drive/MyDrive/capstone/model/label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)


clinicalbert = SentenceTransformer('emilyalsentzer/Bio_ClinicalBERT')
```

```python
def predict_disease(symptoms):
    # Encode the symptoms
    X = clinicalbert.encode([symptoms])

    # Predict
    prediction = model.predict(X, verbose=0)

    # Get disease name
    predicted_class = prediction.argmax()
    confidence = prediction.max()
    disease = label_encoder.inverse_transform([predicted_class])[0]

    return disease, confidence
```

```python
test_cases = [
    "possibly scarring Areas of blackheads Greasy skin Pimples Whiteheads",
]

for symptom in test_cases:
    disease, conf = predict_disease(symptom)
    print(f"{symptom} → {disease} ({conf:.1%})")
```
